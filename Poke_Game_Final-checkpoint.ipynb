```python
"""Welcome to my pokemon game. The storyline follows that of Nintendo's Gameboy SP Pokemon Ruby version where a playe
intially starts the game and procceds to choose his first pokemon if the user names himself Ash and sleeps
through his event he attains an easter egg pikachu otherwise the player procceds to choose 
from the 3 starter types. the the player meets his Rival, May proceeds to battle May and loses his first
battle unless he uses a special move. After which the user proceeds to leave his hometown and head on to his
pokemon Journey. during the phase of coding a lot of hurdles were mer by finding solutions on stack overflow and 
youtube tutorials. The game has a few bugs such as needing specific answers in terms of Y,N or in numbers for specific
choices. """

import time
import datetime
import os

choice= input(prompt="Are you ready to play the best game ever?[Y/N]\n") .upper()
if choice == "N":
    print("lets play another time")

else:
    print("\033[2J\033[0;0H")
    print("\t----------------------------")
    print("\t----------------------------")
    print("\t\033[1;34mPokémon\033[0m \033[1;32m Advanced Battle \033[0m ")
    print("\t----------------------------")
    print("\t----------------------------")
    time.sleep(1)
    print("""\  
    
    _____                      _                    _____ _____  
  / ____|                    | |                  / ____|  __ \ 
 | |  __  __ _ _ __ ___   ___| |__   ___  _   _  | (___ | |__) |
 | | |_ |/ _` | '_ ` _ \ / _ \ '_ \ / _ \| | | |  \___ \|  ___/ 
 | |__| | (_| | | | | | |  __/ |_) | (_) | |_| |  ____) | |     
  \_____|\__,_|_| |_| |_|\___|_.__/ \___/ \__, | |_____/|_|     
                                           __/ |                
                                          |___/                      
    Made in \033[1;33mPython\033[0m Welcome to your childhood!   
                        """)

    
    time.sleep(1)
        ### Start game scenario all variables are initiated###
    def startGame():
        input('Press enter to begin...')

    myName = input('\nWhat\'s your name? ')
    pikachu = '\033[1;33mPikachu\033[0m'
    treeko = '\033[1;32mTreeko\033[0m' 
    torchic = '\033[1;31mTorchic\033[0m' 
    mudkip = '\033[1;36mMudkip\033[0m'
    user_Pokemon = 'CHANGE'
    may_Pokemon = 'CHANGE'
    choicePrompt = '\033[3mchoose a number to act\033[0m'
    choice1 = '\033[1;35m 1.\033[0m'
    choice2 = '\033[1;35m 2.\033[0m'
    choice3 = '\033[1;35m 3.\033[0m'
    choice4 = '\033[1;35m 4.\033[0m'

    print('\nThank you . Best of luck on your pokemon journey!')
    time.sleep(2)
    print("\033[2J\033[0;0H")

    dayOfWeek = datetime.datetime.today().weekday()
    if dayOfWeek == 0:
        day = 'Monday'
    elif dayOfWeek == 1:
        day = 'Tuesday'
    elif dayOfWeek == 2:
        day = 'Wednesday'
    elif dayOfWeek == 3:
        day = 'Thursday'
    elif dayOfWeek == 4:
        day = 'Friday'
    elif dayOfWeek == 5:
        day = 'Saturday'
    elif dayOfWeek == 6:
        day = 'Sunday'



    ### Fail Scenario is initiated when user oversleeps###

    def fail():
        print("""\
        \t*****  \033[1;33m Game Over Please try again \033[0m  *****
        """)
                ### Pokemon center scene is initiaited preceeded by pokemon battle scene###
    def Pokecenter(user_Pokemon):
        time.sleep(1)
        print("\033[2J\033[0;0H")
        print("You arrive at the Pokecenter \n")
        time.sleep(2)
        print(f"Nurse Joy: Hi {myName} I am nurse Joy. Welcome to the Poke Center. I heard about your battle with May let me quickly \n heal your {user_Pokemon} and you should be all set to become a poke master \n")
        count=6
        while count >0:
            count-=1
            print(f"{count} \n")
            time.sleep(1)
            
        print(f"Nurse Joy: {myName} your {user_Pokemon} has been checked,healed and is ready to go on your new adventure") 
        time.sleep(2)
        print(f" {myName} and {user_Pokemon} leave Littleroot Town to start your Poke adventure")
        print("""\
            \t*****  \033[1;34m I hope you Enjoyed the Pokemon Game \033[0m  *****
          """)
                ### Win scenario is initiated if user chooses a specific move###
    def win(user_Pokemon, may_Pokemon):
        time.sleep(1)
        print(f"\n{user_Pokemon} wins!\n")
        time.sleep(1)
        print(f"May and {may_Pokemon} grin and vow to best you one day.Professor Chase congratulates you on your victory.\n")
        time.sleep(1)
        input(f"Press enter to continue...")
        Pokecenter(user_Pokemon)
                ### Lost scenario is initiated when user loses the poke battle###
    def lost(user_Pokemon, may_Pokemon):
        time.sleep(1)
        print(f"\n May's {may_Pokemon} counters with Body Slam, causing {user_Pokemon} to faint. You lose!\n")
        time.sleep(1)
        print(f"May and her {may_Pokemon} Leave with a smug face and vow to never lose to you. Professor Chase congratulates you on your first Poke battle.\n")
        time.sleep(1)
        print(f"Prof. Chase: You and your pokemon will get stronger {myName}, but you must keep training. All the best on your poke adventure\n")
        time.sleep(1)
        input(f"Press enter to continue...")
        Pokecenter(user_Pokemon)
                ### Battle scenario is played out depending on users move choices

    def mayBattle(user_Pokemon, may_Pokemon):
        if user_Pokemon == treeko:
            specialMove = 'Vine Whip'
            winmov = 'Solar Beam'
        elif user_Pokemon == torchic:
            specialMove = 'Flamethrower'
            winmov = 'Overheat'
        elif user_Pokemon == mudkip:
            specialMove = 'Water Gun'
            winmov = 'Aqua Tail'
        time.sleep(1)
        print("\033[2J\033[0;0H")
        print("Battle \n")
        print(f'      {user_Pokemon} versus {may_Pokemon}     \n') 
        time.sleep(1)
        firstAttack = input(f'{choicePrompt} \n {choice1} Tackle \n {choice2} Growl \n {choice3} {specialMove} \n {choice4} {winmov} \n' )
        if firstAttack == '1':
            print(f"{user_Pokemon} tackles May's {may_Pokemon} and inflicts damage. But it isn't very effective")
            lost(user_Pokemon, may_Pokemon)
        if firstAttack == '2':
            print(f"{user_Pokemon} growls at May's {may_Pokemon}.")
            lost(user_Pokemon, may_Pokemon)
        if firstAttack == '3':
            print(f"{user_Pokemon} uses {specialMove} on May's {may_Pokemon}. It's not very effective.")
            lost(user_Pokemon, may_Pokemon)
        if firstAttack == '4':
            print(f"{user_Pokemon} uses {winmov} on Gary's {may_Pokemon}. It's super effective.")
            win(user_Pokemon, may_Pokemon)
            
    ### Pokemon battle scene is initiated###  
    
    def mayChallenge(user_Pokemon, may_Pokemon):
        time.sleep(1)
        print(f'Gary: Before I go become the best trainer and Poke Master, I want my {may_Pokemon} to beat your {user_Pokemon}. Ready, {myName}?\n')
        time.sleep(1)
        print("Entering Pokemon Battle Arena.... Trainer's Get Ready to Battle")
        input(f"\nPress enter to continue...")
        mayBattle(user_Pokemon, may_Pokemon)
        
### Pokemon choosing scene is initiated##

    def mayChooses(user_Pokemon, may_Pokemon):
        time.sleep(1)
        if user_Pokemon == treeko: 
            may_Pokemon = torchic
        elif user_Pokemon == torchic: 
            may_Pokemon = mudkip
        elif user_Pokemon == mudkip: 
            may_Pokemon = treeko
        print(f"May: Haha if {user_Pokemon} is your choice then, I'll take {may_Pokemon} which has a class advantage\n")
        time.sleep(1)
        print(f"May chooses {may_Pokemon}.\n")
        time.sleep(2)
        print(f"May: I & {may_Pokemon} are going to be amazing, We are a much stronger team already.\n")
        mayChallenge(user_Pokemon, may_Pokemon)

    #### choosing pokemon scene is initiated####
    
    def choosePokemon(user_Pokemon):
        choosePoke = input(f'{choicePrompt} \n {choice1} The grass type {treeko} \n {choice2} The fire type {torchic} \n {choice3} The water type {mudkip} \n' )
        if choosePoke == '1':
            user_Pokemon = treeko
        elif choosePoke == '2':
            user_Pokemon = torchic
        elif choosePoke == '3':
            user_Pokemon = mudkip
        time.sleep(1)
        print(f"You've chosen {user_Pokemon}\n")
        mayChooses(user_Pokemon, may_Pokemon)

    ### when user names themselves as Ash a poke easter egg is initiated and they recieve a pikachu####
    def Ash():
        print("\033[2J\033[0;0H")
        print('\nYou arrive at the lab and find the Professor standing by himself\n')
        time.sleep(1)
        print(f'Prof. Chase: What took you so long, {myName}?!')
        time.sleep(1)
        print(f"\nThe three starter Pokemon are gone. No  {treeko}, {torchic}, or {mudkip}.")
        time.sleep(2)
        print('\nBut  wait a minute whats this there is another...\n')
        time.sleep(1)
        print(f'All of a sudden, there is a {pikachu} staring at you with a grin')
        myPokemon = pikachu
        input(f"\nPress enter to continue...")
        Pokecenter(user_Pokemon)
### user decides to sleep in ###
    def late():
        print("\033[2J\033[0;0H")
        print('\nYou arrive at the lab and find the Professor standing by himself\n')
        time.sleep(1)
        print(f'Prof. Chase: What took you so long {myName}!')
        time.sleep(1)
        print(f"\nThe three starter Pokemon are gone. No  {treeko}, {torchic}, or {mudkip}.")
        time.sleep(2)
        print(f'\nSorry... You lose.\n')
        time.sleep(1)
        input(f"\nPress enter to continue...")
        fail()
### poke lab scene is initiated ####
    def labStart():
        print(f'\nYou arrive at the lab and find the professor standing with his grand-daughter.\n')
        time.sleep(1)
        print(f'May: Look who finally decided to show up. Hurry up and choose your pokemon, {myName}.\n')
        time.sleep(1)
        input(f"Press enter to continue...")
        print("\033[2J\033[0;0H")
        time.sleep(1)
        print('\nYou walk up to the professor\'s counter. Whom will you choose?\n')
        time.sleep(2)   
        choosePokemon(user_Pokemon)

    ### wake up svene is inititated####
    def startIntro():
        print('\n\033[3mtring\033[0m')
        time.sleep(1)
        print('\n\033[3mtring tring\033[0m \n ')
        time.sleep(1)
        print(f"Mom: Wake up, {myName}! It's {day}!")
        time.sleep(1)
        print('\nMom: Did you hear me sleepy head?! \n')
        time.sleep(1)

    startIntro()
    ### wake up quickly scene initiated###
    def wakey():
        print("\033[2J\033[0;0H")
        time.sleep(1)
        print("...\n")
        time.sleep(1)
        print(".....\n")
        time.sleep(1)
        print(f"{myName}: I'm up!\n")
        time.sleep(1)
        print("\nMom: Don't be late, today is the day you start of on your poke journey. Professor Chase is waiting for you at his lab.\n")
        time.sleep(1)
        print("You run to the professor's lab. \n")
        time.sleep(1)
        labStart()
    ### sleep in scene is inititated###
    def lazy():
        print("\033[2J\033[0;0H")
        wakeUpResponse2 = input(f'{choicePrompt} \n {choice1} wake up \n {choice2} stay sleeping \n' )
        if wakeUpResponse2 == '1':
            wakey()

        elif wakeUpResponse2 == '2':
            print("\033[2J\033[0;0H")
            print('Dreams of pikachus running around in thier sleep\n')
            time.sleep(1)
            print("...\n")
            time.sleep(1)
            print('THE PROFESSOR!!! \n')
            print('You spring up from your bed and run to the lab in your pajamas')
            if myName == 'Ash':
                Ash()
            else:
                late()
        ### sleep in late scene and seep in fail scene initiated###
    def Asleep():
        print("\033[2J\033[0;0H")
        print("...\n")
        time.sleep(1)
        print(".....\n")
        time.sleep(1)
        print('just few more minutes...\n')
        lazy()

    wakeUpResponse = input(f'{choicePrompt} \n {choice1} wake up \n {choice2} stay sleeping \n' )
    if wakeUpResponse == '1':
        wakey()

    elif wakeUpResponse == '2':
        print("\033[2J\033[0;0H")
        time.sleep(1)
        print('just 5 more minutes please..zz.zz')
        Asleep()


    ### The End###
```
